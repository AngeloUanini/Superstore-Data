# **Tratamiendo de Outliers**


Primero Identificamos los Outliers con el Z-scores

```python
from scipy import stats

# Calcular Z-scores
z_scores = np.abs(stats.zscore(df['sales']))

# Definir un umbral
threshold = 3

# Encontrar los índices de los outliers
outlier_indices_z = np.where(z_scores > threshold)[0]

print("Índices de los outliers usando Z-score:", outlier_indices_z)
```

Vemos el tamaño de los mismos

```python
outlier_indices_z.shape
```

Los vemos en una grafica

Decidimos eliminarlos para evitar problemas ya que al reemplazarlos por la Media podria verse afectada ya que hay bastantes outliers que pueden arruinar la medicion de la misma

```python
df_cleaned = df.drop(outlier_indices_z)

print(f'Tamaño original: {df.shape}')
print(f'Tamaño después de eliminar outliers: {df_cleaned.shape}')
```

# **Arbol de decisión**

Usaremos un modelo de arbol de decision para ver como actua y clasifica nuestros datos

```python
# Importamos las librerias necesarias
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.tree import plot_tree
```

```python
df[['segment', 'category', 'subcategory', 'region', 'city', 'state', 'country']].nunique()
```

```python
# Definir características y variable objetivo
X = df[['discount']].values
y = df['sales'].values
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

```python
# Crear el modelo
model = DecisionTreeRegressor(random_state=42, max_depth=3)

# Entrenar el modelo
model.fit(X_train, y_train)
```

```python
# Hacer predicciones con el conjunto de prueba
y_pred = model.predict(X_test)

# Calcular el error cuadrático medio (MSE) y el coeficiente de determinación (R²)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R² Score: {r2}")
```

```python
# Visualizar el árbol de decisión
plt.figure(figsize=(20, 10))
plot_tree(model, filled=True, feature_names=['discount'], rounded=True)
plt.title('Árbol de Decisión')
plt.show()
```

Podemos ver que al principio separa por discount y luego termina separando por value, siempre por un lado si es mayor y por el otro si es menor.

# **K-Nearest Neighbors**

Usaremos un modelo de clasificacion de vecinos mas cercanos para ver el impacto del descuento en las ventas.


```python
# Importamos las librerias necesarias
from sklearn import neighbors
from sklearn.neighbors import KNeighborsRegressor
from mlxtend.plotting import plot_decision_regions
from sklearn.preprocessing import StandardScaler
```

Elegiremos 'discount' porque queremos analizar su impacto en 'sales', esto permite entender si hay una relación directa y cómo afecta el descuento a las ventas.

Agregamos el 'StandardScaler' para escalar los datos de entrenamiento y prueba.

Hacemos el calculo de metricas para evaluar el modelo de regresion. Estas son el MSE y el R^2.


```python
def knn_comparison(df, k=5):
    # Elegimos las variables
    x = df[['discount']].values
    y = df['sales'].values

    # Dividimos los datos entre test y train
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1234)

    # Escalado de los datos
    scaler = StandardScaler()
    x_train_scaled = scaler.fit_transform(x_train)
    x_test_scaled = scaler.transform(x_test)

    # Creación del modelo KNN Regressor
    clf = KNeighborsRegressor(n_neighbors=k)
    clf.fit(x_train_scaled, y_train)

    # Hacemos la predicción
    y_pred = clf.predict(x_test_scaled)

    # Calcular métricas
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f'MSE: {mse:.2f}')
    print(f'R^2: {r2:.2f}')

    # Gráfico de dispersión para visualizar los resultados
    plt.figure(figsize=(10, 6))
    plt.scatter(x_test, y_test, color='red', label='Datos Reales')
    plt.scatter(x_test, y_pred, color='blue', label='Predicciones')
    plt.xlabel('Descuento')
    plt.ylabel('Ventas')
    plt.title(f'KNN con K = {k}')
    plt.legend()
    plt.show()

# Llamar a la función con el DataFrame y el valor de k deseado
knn_comparison(df, k=5)
```

Podemos ver como de distintos son los datos reales a diferencia de los predichos. Se esperaba que hubiera una cantidad regular de compras en productos sin descuento pero al final parece que no tuvo influencia el descuento en la cantidad de compras.
Lo mismo se ve en los productos con 20% de descuento que no se esperaba gran cantidad de compras y tuvo muchas.
Donde la prediccion acerto fue en el 40%, 45%, e incluso en el 50% ya que las ventas esperadas se acercan a los valores reales, esto se debe a tendencias de compra en productos con este tipo de descuento ya que suelen ser productos muy solicitados que solo estan en oferta en fechas especiales.
Otro acierto es en los productos de 60%, 70% y 80% ya que no se esperan muchas compras de estos ya que suelen ser productos poco solicitados o que se compran muy poco por lo cual se les suele hacer grandes descuentos como estos.
