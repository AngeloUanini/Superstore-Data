# **Arbol de decisión**

Usaremos un modelo de arbol de decision para ver como actua y clasifica nuestros datos

```python
# Importamos las librerias necesarias
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.tree import plot_tree
```

```python
# Convertir variables categóricas en variables dummy (one-hot encoding)
df = pd.get_dummies(df, columns=['segment', 'category', 'subcategory', 'region', 'city', 'state', 'country'])
```

```python
# Seleccionar características y variable objetivo
features = df.drop(columns=['order_id', 'order_date', 'ship_date', 'customer', 'manufactory', 'product_name', 'profit_margin'])
target = df['sales']
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Crear el modelo
model = DecisionTreeRegressor(random_state=42, max_depth=3)

# Entrenar el modelo
model.fit(X_train, y_train)
```

```python
# Hacer predicciones con el conjunto de prueba
y_pred = model.predict(X_test)

# Calcular el error cuadrático medio (MSE) y el coeficiente de determinación (R²)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R² Score: {r2}")
```

```python
# Visualizar el árbol de decisión
plt.figure(figsize=(20,10))
plot_tree(model, filled=True, feature_names=features.columns, rounded=True, fontsize = 10)
plt.show()
```

Podemos ver que al principio separa por cantidad de sales y luego termina separando por value, siempre por un lado si es mayor y por el otro si es menor.

# **K-Nearest Neighbors**

Usaremos un modelo de clasificacion de vecinos mas cercanos para comparar las ventas normales y las ventas con descuento.

```python
# Importamos las librerias necesarias
from sklearn import neighbors
from sklearn.neighbors import KNeighborsRegressor
from mlxtend.plotting import plot_decision_regions
```

```python
def knn_comparison(df, k=5):

    # Definimos las caracteristicas y variable objetivo
    x = df[['discount']].values
    y = df['sales'].values

    # Dividimos los datos entre test y train, ademas de elegir el tamaño de los datos que usara para cada uno
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1234)

    # Usamos Regressor en vez de Classifier ya que son valores continuos | Creacion del modelo
    clf = KNeighborsRegressor(n_neighbors=k)
    clf.fit(x_train, y_train)

    # Hacemos la prediccion
    y_pred = clf.predict(x_test)

    # Gráfico de dispersión para visualizar los resultados
    plt.figure(figsize=(10, 6))
    plt.scatter(x_test, y_test, color='red', label='Datos Reales')
    plt.scatter(x_test, y_pred, color='blue', label='Predicciones')
    plt.xlabel('Descuento')
    plt.ylabel('Ventas')
    plt.title(f'KNN con K = {k}')
    plt.legend()
    plt.show()

# Llamar a la función con el DataFrame y el valor de k deseado
knn_comparison(df, k=5)
```

Podemos ver como de distintos son los datos reales a diferencia de los predichos. Se esperaba que hubiera una cantidad regular de compras en productos sin descuento pero al final parece que no tuvo influencia el descuento en la cantidad de compras.
Lo mismo se ve en los productos con 20% de descuento que no se esperaba gran cantidad de compras y tuvo muchas.
Otra diferencia con la prediccion se ve en los productos con 50% de descuento ya que se esperaba mucho mas volumen de compra y fue mucho menor al esperado, esto se puede deber a que este descuento solo se encuentra en fechas especiales o que los productos con este descuento no son llamativos para los compradores.
