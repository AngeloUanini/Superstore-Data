# Cual es la categoria que mas ventas tiene?
Hipotesis: la categoria de tecnologia deberia tener mas ventas ya que en la actualidad vivimos en un mundo tecnologico y es muy necesaria.

# BarPlot
Primero que nada vamos a ver como se distribuyen las ventas entre las categorias y subcategorias. Para eso vamos a utilizar un grafico de barras.

```python
#definimos el tamaño del grafico
fig, ax = plt.subplots(figsize = (8, 6))
#definimos el tipo de grafico, cuales van a ser sus ejes y que queremos que nos muestre
sns.barplot(x = 'category', y = 'sales', hue = 'subcategory', data = df, estimator = sum)
#definimos los titulos de los ejes y del grafico
plt.title('Distribución de Ventas por Categoría y Subcategoría')
plt.xlabel('Categoría')
plt.ylabel('Ventas Totales')
#sacamos la informacion de las subcateogrias para que no nos tapen el grafico que estamos viendo
plt.legend(title='Subcategoría', bbox_to_anchor=(1.05, 1), loc='upper left')
#mostramos el grafico
plt.show()
```

En el grafico vemos que la que mayor cantidad de ventas termina siendo tecnologia pero por menos diferencia de la que se creia ya que la categoria de muebles tiene una cantidad de ventas similares pero varia mucho mas que tecnologia dependiendo el producto. Y la categoria restante de productos de oficina tiene ventas muy variadas ya que ofrece mas productos.

La Hipotesis es correcta.

# Cual es la region con mejor o mayor margen de ganancia?
Hipotesis: la region con mas habitantes (Oeste) es la que tiene mayor margen de ganancia.

# BoxPlot
Vamos a ver como varia el margen de ganancia dependiendo la region.

```python
#definimos el tamaño del grafico
fig, ax = plt.subplots(figsize = (10, 6))
#definimos cuales van a ser los ejes del grafico
sns.boxplot(x='region', y='profit_margin', data=df)
#definimos los nombres del grafico y los titulos de los ejes
plt.title('Márgenes de Ganancia por Región')
plt.xlabel('Región')
plt.ylabel('Margen de Ganancia')
#mostramos el grafico
plt.show()
```

En el grafico vemos que la region Oeste tiene mas margen de ganancia como se esperaba al tener mayor cantidad de habitantes pero esta mas parejo de lo que se creia, el que llama la atencion es la region central ya que el margen de ganancia no termina siendo significativo ademas de tener valores muy dispersos por lo que se deberia evaluar porque ocurre esto y porque hay tantos valores negativos.

La Hipotesis es correcta.

# ScatterPlot
Usaremos un scatterplot para ver cual es el impacto que tienen los descuentos en las ventas

```python
#definimos el tamaño del grafico
fig, ax = plt.subplots(figsize = (10, 6))
#definimos que columnas usaremos y en donde ademas de el color y la trasparencia de los puntos del grafico
plt.scatter( x = df['discount'], y = df['sales'], alpha = 0.5, c='blue')
#definimos el titulo del grafico y de los ejes
plt.title('Impacto de los Descuentos en las Ventas')
plt.xlabel('Descuento')
plt.ylabel('Ventas')
#ponemos para que se vean las grillas del grafico
plt.grid(True)
#mostramos el grafico
plt.show()
```

En el grafico podemos ver que el impacto en las ventas de los productos en descuento no es lo que se esperaba, no es significativo, esto puede ser ya que los descuentos sean unicamente en fechas especiales (black friday, navidad, etc) o porque los productos en descuento no son llamativos para las personas entonces se ponen en descuento para ver su aumentan las ventas del mismo. De igual forma la mayor cantidad de ventas es en los productos sin descuento.

# Que segmento tiene mas pedidos de los clientes?
Hipotesis: Consumer (consumidor) deberia ser el que tenga mas pedidos ya que es el comprador comun y corriente.

# BarPlot Horizontal
Usaremos un grafico de barras en horizontal para ver quien tiene mas cantidad de pedidos.

```python
#Contamos los pedidos por segmento y ordenamos en 2 columnas
orders_by_segment = df['segment'].value_counts().reset_index()
orders_by_segment.columns = ['segment', 'orders']
#definimos el tamaño del grafico
fig, ax = plt.subplots(figsize=(10, 6))
#definimos el tipo de grafico y el color
plt.barh(orders_by_segment['segment'], orders_by_segment['orders'], color='purple')
#definimos el titulo del grafico y de los ejes
plt.title('Cantidad de Pedidos por Segmento de Cliente')
plt.ylabel('Segmento')
plt.xlabel('Cantidad de Pedidos')
#mostramos el grafico
plt.show()
```

En el grafico vemos que la mayor cantidad de pedidos los hace el consumidor comun, esto se puede deber a que el consumidor hace pedidos mas pequeños e individuales y a que son muchos mas consumidores. En cambio los pedidos corporativos son de empresas y suelen ser pedidos grandes por lo que no son tantos y el home office suele ser de empresas mas chicas, emprendedores o simplemente no son pedidos a nombre de una empresa.

La Hipotesis es correcta.
